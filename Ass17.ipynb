```python
gsv.best_params_
```

```python
model=SVC(C=0.1,gamma=0.001)
model.fit(x_train,y_train)
```

```python
accuracy=accuracy_score(y_test,y_pred) * 100
print(accuracy)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
