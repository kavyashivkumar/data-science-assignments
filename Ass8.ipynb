```python
wine1 = wine.iloc[:, 1:]
wine1
```

```python
wine_array=wine1.values
wine_array
```

```python
wine_norm=scale(wine_array)
wine_norm
```

```python
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1,color='blue')
```

```python
new_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
new_df
```

Hierarchical clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine2=wine.copy()
wine2['clustersid']=hclusters.labels_
wine2
```

k means

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters=KMeans(3,random_state=30).fit(wine_norm)
clusters
```

```python
clusters.labels_
```

```python
wine3=wine.copy()
wine3['clustersid']=clusters.labels_
wine3
```

```python
wine3['clustersid'].value_counts()
```
